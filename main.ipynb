```python
# calculator.py (이 파일 자체가 하나의 '모듈'이 됩니다)

# [함수] 단순 기능
def add(a, b):
    return a + b

# [클래스] 상태와 동작을 가짐
class Calculator:
    def __init__(self, owner):
        self.owner = owner
        self.result = 0

    def accumulate(self, amount):
        self.result += amount
        return f"{self.owner}님의 현재 합계: {self.result}"
```

```python
# 파일명: my_math.py

# 변수 정의
PI = 3.14159

# 함수 정의
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

# 클래스 정의
class Converter:
    def to_fahrenheit(self, celsius):
        return (celsius * 9/5) + 32
```

```python
import my_math

# 1. 모듈 변수 사용
print(f"원주율 값: {my_math.PI}")

# 2. 모듈 함수 호출
result = my_math.add(10, 5)
print(f"더하기 결과: {result}")

# 3. 모듈 클래스 사용
conv = my_math.Converter()
print(f"섭씨 25도는 화씨로 {conv.to_fahrenheit(25)}도입니다.")
```
